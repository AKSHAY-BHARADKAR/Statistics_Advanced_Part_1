```python
# 2.   Generate a Discrete Uniform Distribution using python and Plot the probability mass function (PMF)

import numpy as np
import matplotlib.pyplot as plt

# Set the parameters for the Discrete Uniform Distribution
low, high = 1, 6  # Values between 1 and 6 (inclusive)
size = 1000  # Number of samples to generate

# Generate random samples from the Discrete Uniform Distribution
samples = np.random.choice(range(low, high + 1), size=size, replace=True)

# Calculate the frequency of each outcome
values, counts = np.unique(samples, return_counts=True)

# Calculate the PMF (Probability Mass Function)
pmf = counts / size

# Plot the PMF
plt.bar(values, pmf, color='blue', edgecolor='black')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF) of Discrete Uniform Distribution')

# Display the plot
plt.show()
```

```python
# 3. Write a Python function to Calculate the Probability Distribution Function (PDF) of a Bernoulli Distribution.

def bernoulli_pdf(x, p):
    """
    Calculate the Probability Distribution Function (PDF) of a Bernoulli Distribution.

    Parameters:
    x (int): The outcome of the random variable (either 0 or 1).
    p (float): The probability of success (0 <= p <= 1).

    Returns:
    float: The probability of the outcome x.
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1.")

    # PDF formula for Bernoulli distribution
    return p**x * (1 - p)**(1 - x)

# Example usage:
p = 0.7  # Probability of success (0 <= p <= 1)
x_0 = 0  # Outcome 0
x_1 = 1  # Outcome 1

print(f"PDF for x = 0: {bernoulli_pdf(x_0, p)}")
print(f"PDF for x = 1: {bernoulli_pdf(x_1, p)}")

```

```python
# 9. Write a python function yo Calculate Z-scores from a Dataset and Plot Them.


import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_and_plot_zscores(data):
    """
    Function to calculate Z-scores from a dataset and plot them.
    """
    mean = np.mean(data)       # Compute mean
    std_dev = np.std(data)     # Compute standard deviation
    z_scores = (data - mean) / std_dev  # Compute Z-scores

    # Plot the Z-scores
    plt.figure(figsize=(8, 5))
    sns.histplot(z_scores, bins=30, kde=True, color='purple', edgecolor='black', alpha=0.7)

    # Labels and Title
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.title("Z-score Distribution")
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (Z=0)")
    plt.legend()
    plt.grid()

    # Show the plot
    plt.show()

    return z_scores

# Generate a random dataset (Normal Distribution)
data = np.random.normal(50, 15, 1000)  # Mean=50, Std Dev=15, Sample Size=1000

# Call the function
z_scores = calculate_and_plot_zscores(data)
```

```python
# 14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, data):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    - data_point: The value for which Z-score is computed.
    - data: The dataset used to calculate mean and standard deviation.

    Returns:
    - Z-score of the data point.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_score = (data_point - mean) / std_dev
    return z_score

def plot_standard_normal_distribution(z_score):
    """
    Plot the standard normal distribution and highlight the given Z-score.
    """
    x = np.linspace(-4, 4, 1000)
    pdf_values = norm.pdf(x, 0, 1)  # Standard normal PDF

    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf_values, color='blue', linewidth=2, label="Standard Normal Distribution")

    # Mark the Z-score on the curve
    plt.axvline(z_score, color='red', linestyle='dashed', linewidth=2, label=f"Z-score = {z_score:.2f}")

    # Labels and Title
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.title("Comparison of Z-score with Standard Normal Distribution")
    plt.legend()
    plt.grid()

    # Show the plot
    plt.show()

# Example dataset (random normal data)
data = np.random.normal(50, 10, 1000)  # Mean=50, Std Dev=10, 1000 samples

# Example data point to compute Z-score
data_point = 60

# Compute Z-score
z_score = calculate_z_score(data_point, data)

# Plot the comparison
plot_standard_normal_distribution(z_score)

# Print the computed Z-score
print(f"The Z-score for data point {data_point} is: {z_score:.2f}")
```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import scipy.stats as stats

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test for a given dataset.

    Parameters:
    - sample_data: List or NumPy array of sample observations.
    - population_mean: The known population mean (H0).
    - population_std: The known population standard deviation.
    - alpha: Significance level (default is 0.05 for 95% confidence).

    Returns:
    - Z-score, p-value, and decision (reject or fail to reject H0).
    """
    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Compute the Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Compute the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Decision based on p-value
    if p_value < alpha:
        decision = "Reject the Null Hypothesis (Significant Difference)"
    else:
        decision = "Fail to Reject the Null Hypothesis (No Significant Difference)"

    return z_score, p_value, decision

# Example Dataset (Randomly Generated)
np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=10, size=30)  # Sample mean ~52, Std Dev=10, Sample Size=30

# Population Parameters
population_mean = 50  # Hypothetical population mean
population_std = 10   # Known population standard deviation

# Perform Z-test
z_score, p_value, decision = z_test(sample_data, population_mean, population_std)

# Display Results
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Decision: {decision}")
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Compute the confidence interval for a dataset.

    Parameters:
    - data: List or NumPy array of sample observations.
    - confidence: Confidence level (default is 95%).

    Returns:
    - Tuple (lower bound, upper bound)
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
    sample_size = len(data)

    # Compute the standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Find the critical Z-score
    z_critical = stats.norm.ppf(1 - (1 - confidence) / 2)

    # Compute margin of error
    margin_of_error = z_critical * standard_error

    # Compute confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example Dataset (Randomly Generated)
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Mean ~50, Std Dev=10, Sample Size=30

# Compute 95% Confidence Interval
lower, upper = confidence_interval(sample_data)

# Display Results
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
# 17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def confidence_interval(data, confidence=0.95):
    """
    Compute the confidence interval for a dataset.

    Parameters:
    - data: List or NumPy array of sample observations.
    - confidence: Confidence level (default is 95%).

    Returns:
    - Tuple (lower bound, upper bound)
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
    sample_size = len(data)

    # Compute the standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Find the critical Z-score
    z_critical = stats.norm.ppf(1 - (1 - confidence) / 2)

    # Compute margin of error
    margin_of_error = z_critical * standard_error

    # Compute confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Step 1: Generate data from a normal distribution
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=50)  # Mean=50, Std Dev=10, Sample Size=50

# Step 2: Compute 95% Confidence Interval
lower, upper = confidence_interval(sample_data)

# Step 3: Plot histogram with confidence interval
plt.figure(figsize=(8, 5))
plt.hist(sample_data, bins=10, alpha=0.7, color='blue', edgecolor='black', label="Sample Data")
plt.axvline(lower, color='red', linestyle='dashed', linewidth=2, label=f"Lower CI Bound: {lower:.2f}")
plt.axvline(upper, color='green', linestyle='dashed', linewidth=2, label=f"Upper CI Bound: {upper:.2f}")
plt.axvline(np.mean(sample_data), color='black', linestyle='solid', linewidth=2, label=f"Sample Mean: {np.mean(sample_data):.2f}")

# Labels and legend
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Sample Data with 95% Confidence Interval")
plt.legend()
plt.grid()
plt.show()

# Display CI results
print(f"95% Confidence Interval for the Mean: ({lower:.2f}, {upper:.2f})")
```

```python
# 18. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define mean and standard deviation
mu = 50   # Mean
sigma = 10  # Standard Deviation

# Generate x values (range for the PDF)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)  # Covering 4 standard deviations

# Compute PDF values
pdf_values = norm.pdf(x, mu, sigma)

# Plot the Normal Distribution
plt.figure(figsize=(8, 5))
plt.plot(x, pdf_values, color='blue', linewidth=2, label=f"N({mu}, {sigma}²)")
plt.fill_between(x, pdf_values, alpha=0.3, color='blue')

# Labels and Title
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.title("Probability Density Function (PDF) of a Normal Distribution")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

```python
# 19. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define lambda (average rate of occurrence)
lambda_value = 5  # Average number of occurrences

# Generate possible values for X (event count)
x_values = np.arange(0, 15)  # Compute for 0 to 14 events

# Compute Poisson CDF values
cdf_values = poisson.cdf(x_values, lambda_value)

# Plot the Poisson CDF
plt.figure(figsize=(8, 5))
plt.step(x_values, cdf_values, where='mid', color='blue', label=f"Poisson CDF (λ={lambda_value})")

# Labels and Title
plt.xlabel("Number of Events (k)")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF) of a Poisson Distribution")
plt.legend()
plt.grid()

# Show the plot
plt.show()

# Display probability values
for k, prob in zip(x_values, cdf_values):
    print(f"P(X ≤ {k}) = {prob:.4f}")
```

```python
# 20. Simulate a random variable using a continuous uniform distribution and calculate its expected value.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Define the range [a, b]
a, b = 10, 50  # Lower and upper bounds

# Simulate 1000 random values from Uniform(a, b)
np.random.seed(42)
uniform_data = np.random.uniform(a, b, 1000)

# Calculate the expected value (theoretical and empirical)
expected_value_theoretical = (a + b) / 2
expected_value_empirical = np.mean(uniform_data)

# Plot the histogram
plt.figure(figsize=(8, 5))
plt.hist(uniform_data, bins=30, density=True, color='blue', alpha=0.6, edgecolor='black', label="Simulated Data")

# Plot the theoretical PDF
x = np.linspace(a, b, 1000)
pdf_values = uniform.pdf(x, a, b - a)
plt.plot(x, pdf_values, 'r-', label="Theoretical PDF")

# Labels and legend
plt.axvline(expected_value_theoretical, color='red', linestyle='dashed', linewidth=2, label=f"Theoretical Mean: {expected_value_theoretical:.2f}")
plt.axvline(expected_value_empirical, color='green', linestyle='dashed', linewidth=2, label=f"Empirical Mean: {expected_value_empirical:.2f}")
plt.xlabel("X values")
plt.ylabel("Density")
plt.title("Continuous Uniform Distribution and Expected Value")
plt.legend()
plt.grid()

# Show plot
plt.show()

# Display expected values
print(f"Theoretical Expected Value: {expected_value_theoretical:.2f}")
print(f"Empirical Expected Value from Simulation: {expected_value_empirical:.2f}")
```

```python
# 23. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=500)  # Mean=50, Std Dev=15

# Step 2: Compute mean and standard deviation
mean_original = np.mean(data)
std_original = np.std(data)

# Step 3: Apply Z-score normalization
normalized_data = (data - mean_original) / std_original

# Step 4: Plot Histograms for Comparison
plt.figure(figsize=(10, 5))

# Original Data Histogram
plt.subplot(1, 2, 1)
sns.histplot(data, bins=30, kde=True, color='blue', alpha=0.6)
plt.axvline(mean_original, color='red', linestyle='dashed', linewidth=2, label=f"Mean: {mean_original:.2f}")
plt.title("Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

# Normalized Data Histogram
plt.subplot(1, 2, 2)
sns.histplot(normalized_data, bins=30, kde=True, color='green', alpha=0.6)
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean = 0")
plt.title("Z-Score Normalized Data")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()

# Display calculated values
print(f"Original Mean: {mean_original:.2f}, Standard Deviation: {std_original:.2f}")
print(f"Normalized Mean: {np.mean(normalized_data):.2f}, Standard Deviation: {np.std(normalized_data):.2f}")
```

```python
# 24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def compute_skewness_kurtosis(data):
    """
    Compute and print skewness & kurtosis of a dataset.
    """
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)  # Excess kurtosis (subtracts 3 by default)

    # Print results
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (Excess): {kurtosis:.4f}")
    print(f"Total Kurtosis: {kurtosis + 3:.4f}")  # Adding 3 to get actual kurtosis

    # Plot histogram with KDE
    plt.figure(figsize=(8, 5))
    sns.histplot(data, bins=30, kde=True, color='blue', alpha=0.6)
    plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=2, label="Mean")
    plt.title("Histogram of Generated Data")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid()
    plt.show()

# Generate a normally distributed dataset
np.random.seed(42)
normal_data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10

# Call function
compute_skewness_kurtosis(normal_data)
```
